```python
rootdir = 'rh discography'

def convert_mp3_to_wav_recursive(input_dir):
    for root, dirs, files in os.walk(input_dir):
        for file_name in files:
            if file_name.endswith(".mp3"):
                # Construct full file path
                input_path = os.path.join(root, file_name)
                output_path = os.path.join(root, file_name.replace(".mp3", ".wav"))
                
                # Convert MP3 to WAV
                sound = AudioSegment.from_mp3(input_path)
                sound.export(output_path, format="wav")
                
                print(f"Converted {input_path} to {output_path}")

# Example usage
convert_mp3_to_wav_recursive(rootdir)

#sound = AudioSegment.from_mp3("/path/to/file.mp3")
#sound.export("/output/path/file.wav", format="wav")
```

```python
import os
import librosa
import numpy as np
```

```python
def extract_mel_spectrogram(file_path, n_mels=128, n_fft=2048, hop_length=512):
    y, sr = librosa.load(file_path, sr=None)
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels, n_fft=n_fft, hop_length=hop_length)
    log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)
    return log_mel_spectrogram
```

```python
def process_all_wav_files(input_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for root, dirs, files in os.walk(input_dir):
        for file_name in files:
            if file_name.endswith(".wav"):
                input_path = os.path.join(root, file_name)
                mel_spectrogram = extract_mel_spectrogram(input_path)
                
                # Save the mel-spectrogram as a numpy array
                output_path = os.path.join(output_dir, file_name.replace(".wav", ".npy"))
                np.save(output_path, mel_spectrogram)
                
                print(f"Processed {input_path}, saved mel-spectrogram to {output_path}")


input_wav_dir = 'rh discography' 
output_mel_dir = 'mel_spectrograms'  

# Process all WAV files in the directory
process_all_wav_files(input_wav_dir, output_mel_dir)
```
